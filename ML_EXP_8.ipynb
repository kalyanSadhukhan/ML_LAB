```python
# ------------------------------------------------------------
# Program: Clustering data into groups using K-Means Algorithm
# ------------------------------------------------------------

# Step 1: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Step 2: Load dataset (Iris dataset)
iris = load_iris()
X = iris.data
y = iris.target

# Step 3: Standardize the features (important for K-Means)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 4: Apply K-Means Clustering
k = 3  # Number of clusters (since Iris dataset has 3 species)
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X_scaled)

# Step 5: Get cluster labels
clusters = kmeans.labels_

# Step 6: Add cluster labels to a DataFrame
df = pd.DataFrame(X, columns=iris.feature_names)
df['Cluster'] = clusters

print("âœ… First 10 clustered data points:\n")
print(df.head(10))

# Step 7: Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='rainbow', s=50)
plt.title('K-Means Clustering (Iris Dataset)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Step 8: Compare with actual species
print("\nðŸ“Š Actual vs Predicted Clusters:")
comparison_df = pd.DataFrame({'Actual': y, 'Cluster': clusters})
print(comparison_df.head(10))

# Step 9: Visualize cluster centers
centers = kmeans.cluster_centers_
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='rainbow', s=30)
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.7, marker='X', label='Centroids')
plt.title('Cluster Centers (K-Means)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()

# Step 10: Print cluster centers
print("\nðŸ“ Cluster Centers (in standardized form):\n", centers)
```
